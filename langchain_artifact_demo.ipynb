**Prerequisites:**
- Container Runtime enabled for this notebook
- External Access Integration configured for PyPI access
- Artifact Repository set up (optional, for caching packages)

```python
import langchain
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

print(f"Langchain version: {langchain.__version__}")
```

```python
prompt = PromptTemplate.from_template(
    "Write a SQL query to {task} from the table {table_name}"
)

formatted = prompt.format(
    task="get the top 10 customers by revenue",
    table_name="SALES.CUSTOMERS"
)

print(formatted)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text = """
Snowflake is a cloud-based data warehousing platform that allows users to store, 
analyze, and share data. It provides features like automatic scaling, data sharing, 
and support for semi-structured data. Snowflake's architecture separates storage 
and compute, enabling independent scaling of each layer.
"""

splitter = RecursiveCharacterTextSplitter(
    chunk_size=100,
    chunk_overlap=20
)

chunks = splitter.split_text(text)

for i, chunk in enumerate(chunks):
    print(f"Chunk {i+1}: {chunk}\n")
```
